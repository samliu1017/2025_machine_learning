### Problem 1
Given 
$$
f(x) = \frac{1}{\sqrt{(2\pi)^k |\Sigma|}}
\exp\!\left(-\frac{1}{2}(x-\mu)^T \Sigma^{-1} (x-\mu)\right),
$$  

where $x, \mu \in \mathbb{R}^k$, and $\Sigma$ is a $k\times k$ positive definite matrix with determinant $|\Sigma|$.  
Show that $\int_{\mathbb{R}^k} f(x)\,dx = 1.$  

---
*對角化 \
$\Sigma$ can be diagonalized as  

$$
\Sigma = Q D Q^T,
$$  

where $Q$ is an orthogonal matrix and $D$ is diagonal martix. \
Then, $\Sigma^{1/2} = Q D^{1/2} Q^T$  

so $\Sigma = \Sigma^{1/2}\Sigma^{1/2} $ and $\det(\Sigma^{1/2}) = \sqrt{|\Sigma|}$.  

*變數變換 \
Change the variable 
$$
y = \Sigma^{-1/2}(x - \mu)
\quad \Longrightarrow \quad
x = \mu + \Sigma^{1/2} y.
$$  

And Jacobian determinant is $|\det(\Sigma^{1/2})| = |\Sigma|^{1/2}$,  
so $dx = |\Sigma|^{1/2}\, dy$.  

Then,


$$
\begin{aligned}
\int_{\mathbb{R}^k} f(x)\,dx
&= \frac{1}{\sqrt{(2\pi)^k |\Sigma|}}
   \int_{\mathbb{R}^k}
   \exp\!\left(-\frac{1}{2}(x-\mu)^T \Sigma^{-1} (x-\mu)\right)\,dx \\[6pt]
&= \frac{1}{\sqrt{(2\pi)^k |\Sigma|}}
   \int_{\mathbb{R}^k}
   \exp\!\left(-\frac{1}{2} y^T y\right)
   |\Sigma|^{1/2}\,dy \\[6pt]
&= \frac{1}{(\sqrt{2\pi})^{k}}
   \int_{\mathbb{R}^k} e^{-\frac{1}{2}\sum_{i=1}^k y_i^2}\,dy \\[6pt]
&= \frac{1}{(\sqrt{2\pi})^{k}}
   \prod_{i=1}^k \int_{-\infty}^{\infty} e^{-y_i^2/2}\,dy_i \\[6pt]
&= \frac{1}{(\sqrt{2\pi})^{k}}
   \left(\int_{-\infty}^{\infty} e^{-y^2/2}\,dy\right)^k \\[6pt]
&= \frac{1}{(\sqrt{2\pi})^{k}}
   \left(\sqrt{2\pi}\right)^k \\[6pt]
&= 1
\end{aligned}
$$
** By Calculus ** 

$\int_{-\infty}^{\infty} e^{-y^2/2}\,dy = \sqrt{2\pi},$

Thus, 
$$
\boxed{\displaystyle \int_{\mathbb{R}^k} f(x)\,dx = 1.}
$$

### Problem 2
Let $A,B$ be $n$-by-$n$ matrices and $x$ be a $n$-by-1 vector.

#### (1)Show that $\frac{\partial}{\partial A} \text{trace}(AB) = B^T$

We know
$$
\text{tr}(AB) = \sum_{i=1}^{n} \sum_{k=1}^{n} a_{ik}b_{ki}
$$

The derivative is：

$$
\frac{\partial}{\partial A_{lm}} \sum_k A_{ik} B_{ki} 
= \sum_k \frac{\partial A_{ik}}{\partial A_{lm}} B_{ki}.
= \sum_k \delta_{il} \delta_{km} B_{ki}
= \delta_{il} \sum_k \delta_{km} B_{ki} = \delta_{il} B_{mi}
$$

Thus, 
$$
\frac{\partial}{\partial A_{lm}} \operatorname{trace}(AB) 
= \sum_i \delta_{il} B_{mi} = B_{ml} = B^T.
$$

#### (2)Show that $x^TAx=\text{trace} (xx^TA)$
Left:
$$
\mathbf{x}^{\mathsf{T}}\mathbf{Ax} = \sum_{i=1}^{n} x_i (\mathbf{Ax})_i
= \sum_{i=1}^{n} x_i \left( \sum_{j=1}^{n} A_{ij} x_j \right) = \sum_{i=1}^{n} \sum_{j=1}^{n} x_i A_{ij} x_j
$$


Right:
$$
\text{trace}(\mathbf{xx}^{\mathsf{T}}\mathbf{A}) 
= \sum_{k=1}^{n} (\mathbf{xx}^{\mathsf{T}}\mathbf{A})_{kk}
=  \sum_{k=1}^{n} \sum_{j=1}^{n} (\mathbf{xx}^{\mathsf{T}})_{kj} A_{jk}
= \sum_{k=1}^{n} \left( \sum_{j=1}^{n} x_k x_j A_{jk} \right) = \sum_{k=1}^{n} \sum_{j=1}^{n} x_k A_{jk} x_j
$$

Thus, $x^TAx=\text{trace} (xx^TA)$

#### (3)Derive the maximum likelihood estimators for a multivariate Gaussian.

we have

$$
\ell(\mu, \Sigma)
= -\frac{N k}{2}\ln(2\pi)
-\frac{N}{2}\ln|\Sigma|
-\frac{1}{2}\sum_{i=1}^N (x_i - \mu)^T \Sigma^{-1} (x_i - \mu)
$$
---




$$
\textbf{MLE of } \mu
$$

$$
\ell(\mu) = -\frac{1}{2}\sum_{i=1}^N (x_i-\mu)^T\Sigma^{-1}(x_i-\mu) + \text{const}.
$$

Then, 
$$
\frac{\partial\ell}{\partial\mu}
= -\frac{1}{2}\frac{\partial}{\partial\mu}\left[\sum_{i=1}^N (x_i-\mu)^T\Sigma^{-1}(x_i-\mu)\right]
=-\frac{1}{2}\frac{\partial}{\partial\mu}\left[\sum_{i=1}^N  x_i^T {\Sigma}^{-1} x_i - 2x_i^T \Sigma^{-1} {\mu} + \mu^T \Sigma^{-1} \mu  \right]
= \Sigma^{-1}\sum_{i=1}^N (x_i-\mu).
$$

Let $\frac{\partial\ell}{\partial\mu}=0$,
$$
\Sigma^{-1}\sum_{i=1}^N (x_i-\mu)=0 \quad\Longrightarrow\quad \sum_{i=1}^N (x_i-\mu)=0.
$$

Thus, 
$$
\boxed{\ \hat{\mu}_{\mathrm{MLE}}=\frac{1}{N}\sum_{i=1}^N x_i\ }.
$$

---



$$
\textbf{MLE of } \Sigma
$$


$$
\ell(\Sigma)
= -\frac{N}{2}\ln|\Sigma|
-\frac{1}{2}\sum_{i=1}^N (x_i-\mu)^T\Sigma^{-1}(x_i-\mu)
+ \text{const}.
$$

$$
\frac{\partial \ell}{\partial\Sigma}

-\frac{1}{2}\sum_{i=1}^N \frac{\partial}{\partial\Sigma}(x_i-\mu)^T\Sigma^{-1}(x_i-\mu)
$$


Solve 1. $\frac{\partial ln|\Sigma|}{\partial\Sigma}$

We have

$d(ln|\Sigma|) = \frac{1}{|\Sigma|} d|\Sigma| = tr(\Sigma^{-1} d\Sigma)$

Thus, 
$\frac{\partial ln|\Sigma|}{\partial\Sigma}=(\Sigma^{-1})^T=\Sigma^{-1}$

Solve 2.$\frac{\partial}{\partial\Sigma}(x_i-\mu)^T\Sigma^{-1}(x_i-\mu)$

Because $(x_i-\mu)^T\Sigma^{-1}(x_i-\mu) = tr((x_i-\mu)(x_i-\mu)^T\Sigma^{-1}) = (x_i-\mu)(x_i-\mu)^T$

And proposition $\frac{\partial}{\partial A} f(A^{-1}) = -A^T(\frac {\partial}{\partial A^{-1}}f(A^{-1}))A^{-T}$

Then, $\frac{\partial}{\partial\Sigma}(x_i-\mu)^T\Sigma^{-1}(x_i-\mu) = -\Sigma^{-1}(x_i-\mu)(x_i-\mu)^T\Sigma^{-1}$

So,
$$
\frac{\partial \ell}{\partial\Sigma}
= -\frac{N}{2} \Sigma^{-1}
+\frac{1}{2}\sum_{i=1}^N \Sigma^{-1}(x_i-\mu)(x_i-\mu)^T\Sigma^{-1}
$$


Let $\frac{\partial \ell}{\partial\Sigma} = 0$


$$
\sum_{i=1}^N (x_i-\hat\mu)(x_i-\hat\mu)^T = N\Sigma.
$$

Thus,

$$
\boxed{\ \widehat{\Sigma}_{\mathrm{MLE}}
= \frac{1}{N}\sum_{i=1}^N (x_i-\hat\mu)(x_i-\hat\mu)^T\ }.
$$
