### 1. Show that the sliced score matching (SSM) loss can also be written as

$$ L_{SSM}=\mathbb{E}_{x\sim p(x)} \mathbb{E}_{v\sim p(v)} \left[\|v^TS(x;\theta)\|^2+2v^T\nabla_x (v^TS(x;\theta))\right].$$

---

**從上次作業中SSM loss，繼續推導**

ISM loss:

$$
L_{ISM}(\theta) = \mathbb{E}_{x\sim p(x)}\left[\|S(x;\theta)\|^2 +2\nabla_x\cdot S(x;\theta)\right].$$
where $S(x;\theta) = \nabla_x \log p(x;\theta)$ 

分成兩部分討論 


**第一部分** \
由於假設 $v\sim \mathcal{N}(0,I)$，By Hutchinson’s trace estimator 有

$$
\mathbb{E}_{v\sim p(v)}[vv^T] = I.
$$

所以
$$
|S(x;\theta)\|^2
= S(x;\theta)^T S(x;\theta)
= \mathbb{E}_{v\sim p(v)} \left[S(x;\theta)^T vv^T S(x;\theta)\right]
= \mathbb{E}_{v\sim p(v)} \left[\|v^TS(x;\theta)\|^2\right].
$$ 

**最後得到：**
$\mathbb{E}_{x\sim p(x)}\|S(x;\theta)\|^2 =\mathbb{E}_{x\sim p(x)} \mathbb{E}_{v\sim p(v)}|v^TS(x;\theta)\|^2$


**第二部分**

後項能推論 

$\nabla_x\cdot S(x;\theta)=\sum^d_{i=1}\partial_{x_i}S_i = \operatorname{trace}\!\big(\nabla_x S(x;\theta)\big)$

By Hutchinson’s trace estimator 

$\nabla_x\cdot S(x;\theta)
= \operatorname{trace}\!\big(\nabla_x S(x;\theta)\big)
= \mathbb{E}_{v\sim p(v)}\big[v^\top \big(\nabla_x S(x;\theta)\big) v\big].$

因為$v^\top \big(\nabla_x S(x;\theta)\big) v
= v^\top \nabla_x\big(v^\top S(x;\theta)\big),$

因此$\nabla_x\cdot S(x;\theta)
= \mathbb{E}_{v\sim p(v)}\big[v^\top \nabla_x\big(v^\top S(x;\theta)\big)\big].$

**得到第二部分：**$
\mathbb{E}_{x\sim p(x)}\big[2\nabla_x\cdot S(x;\theta)\big]
= \mathbb{E}_{x\sim p(x)}\mathbb{E}_{v\sim p(v)}\big[2\,v^\top \nabla_x\big(v^\top S(x;\theta)\big)\big].$

最後得出：
$$
L_{SSM}
= \mathbb{E}_{x\sim p(x)}\mathbb{E}_{v\sim p(v)}\Big[||v^\top S(x;\theta)||^2
+2\,v^\top \nabla_x\big(v^\top S(x;\theta)\big)\Big].
$$

### 2. Briefly explain SDE

**隨機微分方程式SDE(stochastic differential equation)** \
是在傳統微分方程中加入隨機噪聲的一種方程，用來描述受不確定性影響的動態系統。

一般形式為：
$dx_t = \underbrace{f(x_t, t)}_{\text{drift}}\,dt + \underbrace{G(x_t, t)}_{\text{diffusion}}\,dW_t, \quad x(0)=x_0,$

當中： \
$f(x_t, t)$ 是漂移項(drift term)：描述系統的**確定性趨勢**。 \
$G(x_t, t)$ 是擴散項(diffusion term)：透過布朗運動引入**隨機性**。
$x_t$隨機過程：代表**系統狀態**，隨時間變化，也是SDE的解 \
$W_t$維納過程：代表**隨機噪音**，也稱為布朗運動 

可透過**伊藤積分**(Itô calculus)求解

---

**例子**：股價隨機運動 (Geometric Brownian Motion) 可以表示為 SDE：

$$
dS_t = \mu S_tdt + \sigma S_tdW_t, \quad S(0) = S_0
$$

其中：
- $\mu$：股價的平均增長率 (drift coefficient)  
- $\sigma$：股價的波動率 (volatility, diffusion coefficient)  
- $W_t$：布朗運動 (Wiener process)

若忽略隨機性 ($\sigma = 0$)，則股價按照確定性增長：
$$
S_t = S_0 e^{\mu t}.
$$

加入隨機性後，股價會在趨勢 $\mu$ 周圍隨機波動，軌跡不可預測。
