我認為20年後的AI可以做到**跨物種的實時溝通與翻譯**

#### 簡化問題提出
我將研究目標設定為「狗」，因為牠們是目前最容易取得相關資料與行為樣本的物種之一。\
本研究的問題為：在限定時間內，能否透過狗的聲音、行為影像或體表溫度變化（以紅外線熱像儀測得）判斷其當下意圖。\
為了簡化問題，先將「意圖」限定為兩種類別──尋求關注與警戒──進行初步分析與模型訓練。

#### 這個簡化問題在概念上如何代表你理想中的最終能力？
此問題的核心在於：AI 是否能在有限時間內準確理解動物當下的意圖。\
若模型能在這個簡化的場景下達成穩定且可靠的判斷，可進一步擴展為更多意圖分類、不同物種間的對應分析，最終發展出能夠雙向溝通、即時互動的跨物種理解模型。

#### 可測試性
為了讓模型更容易學習並達到可靠的二分類效果，我們選取了明顯差異的資料作為訓練集：\
尋求關注的資料：透過多年來狗主人與狗互動的行為紀錄獲得，例如靠近主人、叫聲、搖尾、跳躍等行為。\
警戒的資料：使用已經發生攻擊或威脅人類行為的狗的資料，例如低吼、瞪視、露牙、後退姿勢等行為。\
這樣的資料設計，使兩種類別的行為特徵具有明顯差異，有助於 AI 在監督式學習中更容易判斷狗的當下意圖。\

### Toy Model

想法：輸入聲波以及行為(警戒=1, 尋求關注=0)，透過FFT將波轉成許多頻率，再將頻率丟進nerual network中，最後求出標籤為0, 1，並驗證結果。

將資料切分為：\
Training Set, Validation Set, Test Set:

First:Waveforms -> vectors in $\mathbb{C}^k$(By FFT) -> vectors in $\mathbb{R}^k$(absolute value)

Input: $x \in \mathbb{R}^k$ , Label: $y \in \{0, 1\}$

Output: 0 or 1 (Binary Classification)

Model:  $hθ(x) =σ( W₂ \cdot \text{RELU}(W₁ x + b₁) + b₂ )$, Sigmoid function let output in (0,1)

Loss function: BCE loss

How to check its accurary: Use Confusion Matrix
