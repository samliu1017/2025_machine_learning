### 1.Show that the corresponding probability flow ODE

Since, forward SDE is: 
$$dx_t = f(x_t, t)\,dt + g(x_t, t)dW_t$$
Fokker Planck equtuion is: 
$$\frac{\partial p}{\partial t} = -\frac{\partial}{\partial x}(fp)+ \frac{1}{2}\frac{\partial^2}{\partial x^2}(g^2 p)$$
We want to find deterministic flow $$dx_t = v(x_t, t)dt$$
Then, we get $\frac{\partial p}{\partial t} = -\frac{\partial}{\partial x}(pv)$
$$\frac{\partial p}{\partial t} = -\frac{\partial}{\partial x}(fp)+ \frac{1}{2}\frac{\partial^2}{\partial x^2}(g^2 p) = -\frac{\partial}{\partial x}(pv)$$

$$\frac{\partial}{\partial x}(fp) - \frac{1}{2}\frac{\partial^2}{\partial x^2}(g^2 p) = \frac{\partial}{\partial x}(pv)$$

$$ fp - \frac{1}{2}\frac{\partial}{\partial x}(g^2 p) = pv$$

$$ f - \frac{1}{2p}\frac{\partial}{\partial x}(g^2 p) = v$$

Because,
$$\frac{1}{2p}\frac{\partial}{\partial x}(g^2 p) = \frac{1}{2}(\frac{\partial g^2}{\partial x} + g^2 \frac{1}{p} \frac{\partial p}{\partial x})
= \frac{1}{2}(\frac{\partial g^2}{\partial x} + g^2 \frac{\partial \log p}{\partial x})$$

Then,
$$ dx_t = v(x_t,t)dt = [f(x_t,t) - \frac{1}{2p}\frac{\partial}{\partial x}(g^2 p)]dt =  [f(x_t,t) - \frac{1}{2}\frac{\partial g^2}{\partial x} - \frac{1}{2}g^2 \frac{\partial \log p}{\partial x}]dt$$

$$ dx_t =  [f(x_t,t) - \frac{1}{2}\frac{\partial}{\partial x}g^2(x_t, t) - \frac{g^2(x_t, t)}{2}\frac{\partial}{\partial x} \log p(x_t,t)]dt$$
